#### Zomato has more working in Indian and has more transactions in India.

```python
country_names= final_df['Country'].value_counts().index
```

```python
country_names = np.array(country_names)
```

```python
country_names
```

```python
country_values= final_df['Country'].value_counts().values
```

```python
country_values
```

```python
plt.figure(figsize=[20,12])
plt.bar(country_names,country_values,color= 'Maroon')

plt.title("Country wise transactions")
     
plt.xlabel("Country")
plt.ylabel("Number of transactions")
plt.show()
```

```python
plt.pie(x = country_values[:3],labels= country_names[:3],autopct='%1.2f%%')
plt.show()
```
